```python
class Node :
    def __init__(self, val, priority) :
        self.val = val
        self.priority = priority
        
    def __repr__(self) :
        return str(self.__dict__)
```

```python
class Priority_Queue :
    
    def __init__(self) :
        self.values = []
    
    def enqueue(self, val, priority) :
        new_node = Node(val, priority)
        self.values.append(new_node)
        if len(self.values) > 1:
            self.bubble_up()
        return self.values

    def bubble_up(self) :
        child_idx = len(self.values)-1
        
        while child_idx > 0 :
#             if child == 0 : # single node
#                 break

            parent_idx = (child_idx-1)//2 # floor value

            child_node = self.values[child_idx]
            parent_node = self.values[parent_idx]
            if parent_node.priority < child_node.priority :
                self.values[parent_idx], self.values[child_idx] = self.values[child_idx], self.values[parent_idx]
                child_idx = parent_idx
                parent_idx = (child_idx-1)//2 # floor value
#                print(self.values)

            else :
                break
            
    def dequeue(self) :
        if len(self.values) == 0 :
            return None
            
        self.values[0], self.values[-1] = self.values[-1], self.values[0]
        rem_elem = self.values.pop()
        if len(self.values) > 1 :
            self.bubble_down()
        return rem_elem
    
    def bubble_down(self) :
        parent_idx = 0
        
        while True :
            swap = None
            child_idx_1 = 2*parent_idx + 1
            child_idx_2 = 2*parent_idx + 2

            parent_node = self.values[parent_idx]

            # if two child exists
            if (child_idx_2) < len(self.values) :
                
                child_1_node = self.values[child_idx_1]
                child_2_node = self.values[child_idx_2]
                
                if child_1_node.priority > child_2_node.priority :
                    max_child = child_1_node
                else :
                    max_child = child_2_node

#               max_child = max(child_1, child_2)
                if max_child.priority > parent_node.priority :
                    swap = self.values.index(max_child) # swap is containing int index values
            # if one child exists
            elif (child_idx_1) < len(self.values) :
                
                child_1_node = self.values[child_idx_1]
                if parent_node.priority < child_1_node.priority :
                    swap = child_idx_1
            
            print("Parentidx = ", parent_idx, "swap_idx = ", swap, "Values len = ", len(self.values))
            if swap is None :
                break
                
            self.values[parent_idx], self.values[swap] = self.values[swap], self.values[parent_idx]
            parent_idx = swap
    
#       [200, 100, 36, 17, 19]
```
